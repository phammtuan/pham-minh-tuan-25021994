```python
def primenum(n):
    a=2
    if n==2:
        return True
    elif n!=2: 
        while a<n:
            if n%a!=0:
                a=a+1
                return True
            elif n%a==0:
                return False
b=int(input())
primenum(b)
    
```

```python
def uocso(n):
    c=[]
    b=1
    while b<n:
        if n%b==0:
            c.append(b)
        b=b+1
    return(c)
def perfectnum(a):
    if sum(uocso(a))==a:
        return True
    else:
        return False
m=int(input())
perfectnum(m)

```

```python
def calculator(num1, num2, operat):
    if operat == '+':
        return round(num1 + num2, 2)
    elif operat == '-':
        return round(num1 - num2, 2)
    elif operat == '*':
        return round(num1 * num2, 2)
    elif operat == '/':
        if num2 == 0:
            return "Cannot divide by zero"
        return round(num1 / num2, 2)
    else:
        return "Invalid operator"
num1,op,num2=map(str,input().split())
num1=float(num1)
num2=float(num2)
calculator(num1,num2,op)

```

```python
def kiem_tra_dang_cau(a, b):
    map_a_to_b = {}
    map_b_to_a = {}
    
    for char_a, char_b in zip(a, b):
        
        if char_a in map_a_to_b:
            if map_a_to_b[char_a] != char_b:
                return False
        else:
            map_a_to_b[char_a] = char_b
        if char_b in map_b_to_a:
            if map_b_to_a[char_b] != char_a:
                return False
        else:
            map_b_to_a[char_b] = char_a
    return True
m,n=map(str,input().split())
kiem_tra_dang_cau(m,n)
```
