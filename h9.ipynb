```python
def is_path_crossing(moves) :
    x = 0 
    y = 0
    path = [[0,0]]
    for i in moves :
        if i == 'R' :
            x += 1
            if [x,y] in path : return 'True'
            else :
                path.append([x,y])
        elif i == 'L' :
            x -= 1
            if [x,y] in path : return 'True'
            else :
                path.append([x,y])
        elif i == 'U' :
            y += 1
            if [x,y] in path : return 'True'
            else :
                path.append([x,y])
        elif i == 'D' :
            y -= 1
            if [x,y] in path : return 'True'
            else :
                path.append([x,y])
    return 'False'

moves = input()

print(is_path_crossing(moves))
```

```python
def Median(nums) :
    a,b,c,d,e = nums 
    if a < b : b1 = [a,b] # 1
    else : b1 = [b,a]
    if c < d : b2 = [c,d] # 2
    else : b2 = [d,c]
    if b1[1] < b2[1] : # 3
        b3 = [b2[0]] + [e]
        b4 = b1
        
    else :
        b3 = [b1[0]] + [e]
        b4 = b2 
    if b3[0] > b3[1] : b3[0] , b3[1] = b3[1] , b3[0] # 4

    if b4[1] < b3[1] : # 5 
        if b4[1] < b3[0] : return b3[0] # 6
        else : return b4[1]
    else :
        if b4[0] > b3[1] : return b4[0]
        else : return b3[1]


nums = list ( map ( int , input().split() ) )
print(Median(nums))
```

```python
def binary_search(arr,target) :
    left , right = 0 , len(arr) - 1
    r_min = []
    while left <= right :
        mid = (left + right) // 2 
        if  arr[mid] == target :
            return 0 
        elif arr[mid] < target :
            left  =  mid + 1
            r_min.append(abs(arr[mid] - target)) 
        else :
            right = mid - 1
            r_min.append(abs(arr[mid] - target)) 
    return min(r_min) 

def  findRadius(houses, heaters) :
    houses.sort()
    heaters.sort()
    radius = []
    for i in houses :
        radius.append(binary_search(heaters,i))
    return max(radius)



houses = list ( map ( int , input().split() ) )
heaters = list ( map ( int , input().split() ) )

print(findRadius(houses,heaters))
```
