```python
def dem_so_tu(chuoi):
    so_tu = 0
    
    for i in range(len(chuoi)):
        ky_tu = chuoi[i]
        
        if 'a' <= ky_tu <= 'z' or 'A' <= ky_tu <= 'Z':
            if i == 0 or chuoi[i-1] == ' ':
                so_tu += 1
                
    return so_tu

chuoi_nhap = input("Nhập chuỗi (chỉ chứa chữ cái và khoảng trắng): ")
print(f"Số từ trong chuỗi là: {dem_so_tu(chuoi_nhap)}")
```

```python
def in_tu_dau_tien(chuoi):
    tu_dau_tien = ""
    start_index = -1
    
    for i in range(len(chuoi)):
        if chuoi[i] != ' ':
            start_index = i
            break
            
    if start_index == -1:
        return "Chuỗi không có từ nào."
        
    for i in range(start_index, len(chuoi)):
        ky_tu = chuoi[i]
        
        if ky_tu == ' ':
            break
        else:
            tu_dau_tien += ky_tu
            
    return tu_dau_tien

chuoi_nhap = input("Nhập chuỗi: ")
print(f"Từ đầu tiên trong chuỗi là: {in_tu_dau_tien(chuoi_nhap)}")
```

```python
def tinh_tong_3_so(chuoi):
    tong = 0
    so_hien_tai = ""
    
    chuoi_xu_ly = chuoi + "," 
    
    for i in range(len(chuoi_xu_ly)):
        ky_tu = chuoi_xu_ly[i]
        
        if '0' <= ky_tu <= '9':
            so_hien_tai += ky_tu
        elif ky_tu == ',':
            if so_hien_tai:
                tong += int(so_hien_tai) 
            
            so_hien_tai = ""
            
    return tong

chuoi_nhap = input("Nhập chuỗi (VD: 3, 12, 15): ")
print(f"Tổng 3 số nguyên là: {tinh_tong_3_so(chuoi_nhap)}")
```

```python
def dem_ky_tu(chuoi):
    dem_hoa = 0
    dem_thuong = 0
    dem_so = 0
    
    for i in range(len(chuoi)):
        ky_tu = chuoi[i]
        
        ascii_val = ord(ky_tu)
        
        if 65 <= ascii_val <= 90:
            dem_hoa += 1
        
        elif 97 <= ascii_val <= 122:
            dem_thuong += 1
            
        elif 48 <= ascii_val <= 57:
            dem_so += 1
            
    print(f"Số ký tự In hoa: {dem_hoa}")
    print(f"Số ký tự In thường: {dem_thuong}")
    print(f"Số ký tự Số: {dem_so}")

chuoi_nhap = input("Nhập chuỗi: ")
dem_ky_tu(chuoi_nhap)
```

```python
def tinh_tong_so(chuoi):
    tong = 0
    
    print("Các ký tự số đã tách:", end=" ")
    
    for i in range(len(chuoi)):
        ky_tu = chuoi[i]
        
        if '0' <= ky_tu <= '9':
            gia_tri_so = int(ky_tu)
            tong += gia_tri_so
            print(gia_tri_so, end=" + ")
            
    print("\b\b= ", end="") 
    return tong

chuoi_nhap = input("Nhập chuỗi (VD: abd45ecf47wde3s1): ")
tong_cac_so = tinh_tong_so(chuoi_nhap)
print(tong_cac_so)
```

```python
def kiem_tra_mat_khau_manh(mat_khau):
    co_ky_tu_dac_biet = False
    co_in_hoa = False
    co_con_so = False
    co_chu_thuong = False
    do_dai_hop_le = (len(mat_khau) > 6)

    for i in range(len(mat_khau)):
        ky_tu = mat_khau[i]
        ascii_val = ord(ky_tu)

        if 65 <= ascii_val <= 90:
            co_in_hoa = True
        
        elif 97 <= ascii_val <= 122:
            co_chu_thuong = True
            
        elif 48 <= ascii_val <= 57:
            co_con_so = True
            
        else:
            co_ky_tu_dac_biet = True
            
        if co_in_hoa and co_chu_thuong and co_con_so and co_ky_tu_dac_biet:
            break
            
    if (co_ky_tu_dac_biet and co_in_hoa and co_con_so and co_chu_thuong and do_dai_hop_le):
        return "Mật khẩu MẠNH."
    else:
        print("Mật khẩu YẾU vì thiếu các điều kiện sau:")
        if not do_dai_hop_le: print(" - Độ dài phải lớn hơn 6")
        if not co_ky_tu_dac_biet: print(" - Thiếu 1 ký tự đặc biệt")
        if not co_in_hoa: print(" - Thiếu 1 ký tự in hoa")
        if not co_con_so: print(" - Thiếu 1 con số")
        if not co_chu_thuong: print(" - Thiếu 1 chữ thường")
        return "Mật khẩu KHÔNG MẠNH."

mat_khau_nhap = input("Nhập chuỗi mật khẩu: ")
print(kiem_tra_mat_khau_manh(mat_khau_nhap))
```

```python
def format_so_phan_cach_ngan(so_nguyen):
    chuoi_so = str(so_nguyen)
    chuoi_ket_qua = ""
    dem_ky_tu = 0
    
    for i in range(len(chuoi_so) - 1, -1, -1):
        ky_tu = chuoi_so[i]
        
        if dem_ky_tu == 3:
            chuoi_ket_qua = "." + chuoi_ket_qua
            dem_ky_tu = 0
        
        chuoi_ket_qua = ky_tu + chuoi_ket_qua
        dem_ky_tu += 1
            
    return chuoi_ket_qua

try:
    so_nhap = int(input("Nhập vào một số nguyên: "))
    print(f"Đổi sang chuỗi rồi in ra: {format_so_phan_cach_ngan(so_nhap)}")
except ValueError:
    print("Lỗi: Đầu vào không phải là số nguyên hợp lệ.")
```
