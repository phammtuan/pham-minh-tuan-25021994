```python
import math
with open("input.txt", "r") as file: 
    s = file.readlines()
    lst = [lst.strip() for lst in s if lst.strip()]
    for i in range(0,len(lst),2):
        type = lst[i]
        data = lst[i+1]
        shape = type.split(":")[1].strip().upper() # lấy tên loại hình

        if shape == "SQUARE": 
            side = data.split(":")[1]
            print( f"{4 * float(side):.2f}")    
        elif shape == "RECTANGLE": 
            side = data.replace(":", " ")
            parts = side.split() 
            width = float(parts[1])
            height = float(parts[3])
            print(f"{2 * (width + height):.2f}")
        elif shape == "CIRCLE": 
            radius = data.split(":")[1]
            print(f"{2 * math.pi*float(radius):.2f}")
```

```python
def power(arr, i): 
    tem = 1 
    for j in range(0,len(arr)):
        if j == i:
            continue
        else: 
            tem *= arr[j]
    return tem
with open("input.txt", "r") as file: 
    n = file.readline()
    s = file.readline()
    s = s.split() # always have this step 
    lst = [int(n) for n in s]
    res = []
    for i in range(0,len(lst)): 
        res.append(power(lst,i))
    print(res)
```

```python
import math

try:
    with open("input.txt", "r") as file:
        s = file.read().split()
        M = int(s[0])
        raw = [int(x) for x in s[1:]]
        img = []
        for i in range(M):
            img.append(raw[i*M : (i+1)*M])
        res = [[0] * M for _ in range(M)]
        for r in range(M):
            for c in range(M):
                sum = 0
                count = 0
                for i in range(r - 1, r + 2):
                    for j in range(c - 1, c + 2):
                        if 0 <= i < M and 0 <= j < M:
                            sum += img[i][j]
                            count += 1
                res[r][c] = sum // count
        for row in res:
            print(*(row))
            
except:
    print("Mission failed")
```
