```python
def search(a,val):
    
    for i in a:
        if i==val:
            return a.index(i)
    return '-1'
n=list(map(int,input().split()))
k=int(input())
search(n,k)
```

```python
def count(lis,target):
    j=0
    for i in lis:
        if i==target:
            j+=1
    return j
nums=list(map(int,input().split()))
num=int(input())
count(nums,num)
```

```python
def bubblesort(nums):
    for i in range(len(nums)):
        swapped=False
        for j in range(0,len(nums)-i-1):
            if nums[j]>nums[j+1]:
                nums[j],nums[j+1]=nums[j+1],nums[j]
                swapped=True
        if not swapped:
            break
    return nums
a=list(map(int,input().split()))
bubblesort(a)
```

```python
def selectionsort(nums):
    for i in range(0,len(nums)):
        for j in range(i+1,len(nums)):
            if nums[j]<nums[i]:
                nums[j],nums[i]=nums[i],nums[j]
    return nums
a=(list(map(int,input().split())))
selectionsort(a)

```

```python
def binary(n,k):
    left=0
    right=len(n)-1
    while left<right:
        mid =(left+right)//2
        if n[mid]==k:
            return mid
        else:
            if n[mid]<k:
                left=mid+1
            elif n[mid]>k:
                right=mid-1
n=int(input())
guests=[]
for i in range(n):
    guest=input()
    guests.append(guest)
index=binary(guests,'Nemo')
if index==0:
    print(guests[1])
elif index==len(guests)-1:
    print(guests[-2])
else:
    print(guests[index-1],'and',guests[index+1])
    
```
