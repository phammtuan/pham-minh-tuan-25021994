```python
class Cylinder:
    def __init__(self,r,h):
        self.r=r
        self.h=h
    def thetich(self):
        return (self.r**2)*3.14*self.h
    def dientich(self):
        return 2*(self.r**2)*3.14+2*self.r*3.14*self.h
bankinh=float(input())
chieucao=float(input())
xylanh=Cylinder(bankinh,chieucao)
print(xylanh.dientich())
print(xylanh.thetich())
        
    
```

```python
class Calculator:
    def __init__(self, a, b):
        """Khởi tạo với 2 số a và b"""
        self.a = a
        self.b = b

    def add(self):
        """1. Cộng"""
        return self.a + self.b

    def subtract(self):
        """2. Trừ"""
        return self.a - self.b

    def product(self):
        """3. Nhân"""
        return self.a * self.b

    def divide(self):
        """4. Chia"""
        if self.b == 0:
            return "Lỗi: Không thể chia cho 0"
        return self.a / self.b

    def power(self):
        """5. Lũy thừa (Mũ)"""
        return self.a ** self.b

    def mod(self):
        """6. Chia lấy phần dư"""
        if self.b == 0:
            return "Lỗi: Không thể chia cho 0"
        return self.a % self.b

    def set_numbers(self, a, b):
        """7. Cập nhật 2 số mới"""
        self.a = a
        self.b = b
        print(f"Đã cập nhật số thành công: {self.a} và {self.b}")


def main():
    print("CHƯƠNG TRÌNH MÁY TÍNH BỎ TÚI")
    
    try:
        num1 = float(input("Nhập số thứ nhất: "))
        num2 = float(input("Nhập số thứ hai: "))
    except ValueError:
        print("Vui lòng nhập số hợp lệ.")
        return

    
    calc = Calculator(num1, num2)

    while True:
        print("\n--- MENU ---")
        print(f"Hai số hiện tại: {calc.a} và {calc.b}")
        print("1. Cộng (add)")
        print("2. Trừ (subtract)")
        print("3. Nhân (product)")
        print("4. Chia (divide)")
        print("5. Lũy thừa (power)")
        print("6. Chia lấy dư (mod)")
        print("7. Nhập cặp số mới (set_numbers)")
        
        choice = input("Chọn chức năng (1-7): ")

        result = None

        if choice == '1':
            result = calc.add()
            print(f"Kết quả cộng: {result}")
        elif choice == '2':
            result = calc.subtract()
            print(f"Kết quả trừ: {result}")
        elif choice == '3':
            result = calc.product()
            print(f"Kết quả nhân: {result}")
        elif choice == '4':
            result = calc.divide()
            print(f"Kết quả chia: {result}")
        elif choice == '5':
            result = calc.power()
            print(f"Kết quả lũy thừa: {result}")
        elif choice == '6':
            result = calc.mod()
            print(f"Kết quả chia dư: {result}")
        elif choice == '7':
            try:
                n1 = float(input("Nhập số mới thứ nhất: "))
                n2 = float(input("Nhập số mới thứ hai: "))
                calc.set_numbers(n1, n2)
            except ValueError:
                print("Lỗi: Vui lòng nhập số hợp lệ.")
        else:
            print("Lựa chọn không hợp lệ, vui lòng chọn lại.")

        
        check_exit = input("\nBạn có muốn thoát? ")
        if check_exit in ["yes", "YES", "Yes"]:
            print("Đã thoát chương trình.")
            break

if __name__ == "__main__":
    main()
```

```python
class Rectangle:
    def __init__(self,height,width):
        self.height=height
        self.width=width
    def area(self):
        return 'dien tich la:', self.height*self.width
    def perimeter(self):
        return 'chu vi la:', 2*(self.height+self.width)
    def scale(self):
        k=float(input())
        self.height=self.height*k
        self.width=self.width*k
a,b=map(float,input().split())
hinh1=Rectangle(a,b)
print(hinh1.scale())
print(hinh1.area())
```

```python
class Triangle:
    def __init__(self,edge1,edge2,edge3):
        self.edge1=edge1
        self.edge2=edge2
        self.edge3=edge3
    def check(self):
        if self.edge1 + self.edge2<=self.edge3:
            return False
        elif self.edge2 + self.edge3<=self.edge1:
            return False
        elif self.edge1+self.edge3<=self.edge2:
            return False
        return True
    def area(self):
        p=(self.edge1+self.edge2+self.edge3)/2
        ans=(p*(p-self.edge1)*(p-self.edge2)*(p-self.edge3))**0.5
        return ans
a,b,c=map(int,input().split())
tam1=Triangle(a,b,c)
if tam1.check():
    dien=tam1.area()
    print(f'{dien:.2f}')
else:
    print('Invalid')
```

```python
class Bankaccount:
    def __init__(self,owner,balance):
        self.owner=owner
        self.balance=balance
    def deposit(self,nums):
        self.balance=int(self.balance)+nums
    def withdraw(self,nums):
        if self.balance>=nums:
            self.balance=int(self.balance)-nums
        else:
            self.balance=self.balance
cus=input()
cus=cus.split()
cus=Bankaccount(cus[0],cus[1])
k=int(input())
for i in range(k):
    order=input()
    order=order.split()
    if order[0]=='deposit':
        cus.deposit(int(order[1]))
    else:
        cus.withdraw(int(order[1]))
print(cus.balance)
    
    
```

```python
class ShoppingCart:
    def __init__(self):
        """1. Khởi tạo với danh sách rỗng"""
        self.products = []

    def add_product(self, name, price):
        """2. Thêm vào giỏ hàng một sản phẩm mới"""
        product = {"name": name, "price": price}
        self.products.append(product)
        print(f"--> Đã thêm '{name}' với giá {price} vào giỏ hàng.")

    def remove_product(self, name):
        """3. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên)"""
        found = False
        for i, product in enumerate(self.products):
            if product["name"] == name:
                del self.products[i]
                print(f"--> Đã xóa sản phẩm '{name}' khỏi giỏ hàng.")
                found = True
                break # Chỉ xóa sản phẩm đầu tiên tìm thấy
        
        if not found:
            print(f"--> Không tìm thấy sản phẩm tên '{name}' trong giỏ.")

    def check_empty(self):
        """4. Kiểm tra xem giỏ hàng có rỗng không"""
        if len(self.products) == 0:
            print("--> Giỏ hàng hiện đang TRỐNG.")
            return True
        else:
            print("--> Giỏ hàng KHÔNG trống.")
            return False

    def calculate_total(self):
        """5. Tính tổng tiền giỏ hàng"""
        total = sum(p["price"] for p in self.products)
        print(f"--> Tổng tiền giỏ hàng là: {total}")
        return total

    def show_cart(self):
        """6. Hiển thị toàn bộ sản phẩm trong giỏ"""
        if len(self.products) == 0:
            print("--> Giỏ hàng trống, không có gì để hiển thị.")
        else:
            print("\n--- DANH SÁCH SẢN PHẨM ---")
            for index, product in enumerate(self.products, 1):
                print(f"{index}. Tên: {product['name']} - Giá: {product['price']}")
            print("--------------------------")

    def clear_cart(self):
        """7. Xóa toàn bộ giỏ hàng"""
        self.products = []
        print("--> Đã xóa toàn bộ giỏ hàng.")

# --- CHƯƠNG TRÌNH CHÍNH (MAIN) ---

def main():
    cart = ShoppingCart()

    while True:
        print("\n================ MENU ================")
        print("2. Thêm sản phẩm vào giỏ hàng")
        print("3. Xóa sản phẩm khỏi giỏ hàng")
        print("4. Kiểm tra giỏ hàng rỗng")
        print("5. Tính tổng tiền")
        print("6. Hiển thị giỏ hàng")
        print("7. Xóa toàn bộ giỏ hàng")
        print("======================================")
        
        choice = input("Nhập lựa chọn của bạn (2-7): ")

        if choice == '2':
            name = input("Nhập tên sản phẩm: ")
            try:
                price = float(input("Nhập giá sản phẩm: "))
                cart.add_product(name, price)
            except ValueError:
                print("Lỗi: Giá phải là một con số.")
        
        elif choice == '3':
            name = input("Nhập tên sản phẩm cần xóa: ")
            cart.remove_product(name)
            
        elif choice == '4':
            cart.check_empty()
            
        elif choice == '5':
            cart.calculate_total()
            
        elif choice == '6':
            cart.show_cart()
            
        elif choice == '7':
            cart.clear_cart()
            
        else:
            print("Lựa chọn không hợp lệ. Vui lòng chọn từ 2 đến 7.")

        # Yêu cầu thoát chương trình theo đề bài
        print() # Dòng trống cho dễ nhìn
        exit_q = input('Có muốn thoát khỏi chương trình không? (nhập "yes" để thoát): ')
        if exit_q.lower() == "yes":
            print("Cảm ơn bạn đã sử dụng chương trình!")
            break

if __name__ == "__main__":
    main()
```
