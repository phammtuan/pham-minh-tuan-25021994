```python
def happy(n):
    ans=0
    while n!=0:
        ans =(n%10)**2+ans
        n=n//10
    return ans
k=int(input())
while True:  
    if happy(k)==1:
        print('True')
        break
    elif happy(k) != 4:
        k=happy(k)
    elif happy(k) ==4:
        print('False')
        break
```

```python
try: 
    a = list(map(int, input().split()))
    dct = {}
    sum = 0
    check = True
    for i in range(1, len(a)): 
        if dct.get(a[i], 0) == 0:
            sum += a[i]
            dct[a[i]] = 1
        else: 
            check = False
            break
    if check : print(sum)
    else: print("Mang khong hop le")
except ValueError: 
    print("Mang khong hop le")
except IndexError: 
    print("Mang khong dap ung so luong phan tu")
```

```python
import re
import sys
n = int(input())
while n > 0:
    s = input().strip() 
    try:
        re.compile(s)
        print("Valid")
    except re.error: 
        print("Invalid")
    n -= 1
```
